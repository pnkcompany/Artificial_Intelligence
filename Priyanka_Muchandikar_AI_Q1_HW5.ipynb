```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

ds=np.array([[1,0,0,7,'Apple'],
             [0,1,0,20,'Watermelon'],
             [1,0,0,1,'Cherry'],
             [0,1,0,7.5,'Apple'],
             [1,0,0,1,'Strawberry'],
             [1,0,0,0.8,'Cherry']])
print(ds)
X=ds[:,:4]
y=ds[:,4]

rdmfr=RandomForestClassifier(bootstrap=True,random_state=5,max_features='sqrt',max_depth=2,n_estimators=200,criterion='gini')
rdmfr.fit(X,y)
tst=np.array([[0,1,1,0.4]])
rdmfr.predict(tst)

est=rdmfr.estimators_[5]
plt.figure()
plott=plot_tree(est,feature_names=['Green','Yellow','Red','Pink'],
                class_names=['Apples','Watermelon','Cherry','Strawberry'],
                filled=True,
                fontsize=10)
```
